#### 07. Rename the column names

```python
df = df.rename(columns = {'Player':'Player_Name',
                          'Span':'Duration_of_Career',
                          'Mat':'No_of_match_played', 
                          'Inns':'No_of_innings', 
                          'Balls':'No_of_balls_bowled',
                          'Runs': 'Total_Runs_conceded', 
                          'Wkts':'Total_wickets_taken', 
                          'BBI': 'Best_Bowling_in_an_Innings', 
                          'BBM':'Best_Bowling_in_a_Match',
                          'Ave':'Bowling_Average',
                          'Econ':'average_runs_conceded_per_over', 
                          'SR':'Strike_rate',
                          5: 'No_of_times_5_wickets_taken', 
                          10:'No_of_times_10_wickets_taken'})
display(df.head())
```
