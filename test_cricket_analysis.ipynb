## Homework 02

#### 01. Removing the columns BBI and BBM.

```python
data=df.copy()
data=data.drop(['BBI', 'BBM'],  axis=1) # Drop 'BBI' and 'BBM'
display(data.head(10))
```

#### 02. How many players played for ICC?

```python
# splitting the 'Player' column to get the information about 'Country'
data_player = data['Player'].str.split("(", expand=True)

# concatinating 'Country' with the main dataframe
data = pd.concat([data, data_player], axis=1)

# dropping the 'Player' columns
data = data.drop('Player', axis=1)

# renaming the column names
data = data.rename(columns={0: 'Player',
                        1: 'Country'})
new_columns=['Player','Span', 'Mat', 'Inns','Balls', 'Runs','Wkts', 'Ave','Econ',
                   'SR',5, 10,'Country']
data=data[new_columns]
# remove the ")" from the 'Country' column
data['Country'] = data['Country'].str.replace(")", "")

# create a column based on a condition or function
def icc_check(x):
    if "ICC" in x:
        return "Yes"
    else:
        return "No"

##Finding number of player played for ICC    
data['Played_for_ICC'] = data['Country'].apply(icc_check)
print(data['Played_for_ICC'].value_counts())
```

A total 5 number of players played for ICC.

#### 03. How many different countries are present in this dataset? 

```python
# Removing "ICC/" from the 'Country'
data['Country']=data['Country'].str.replace("ICC/", "")
data['Country']=data['Country'].str.replace("/ICC", "")
data['Country'].value_counts()
```

10 different countries are present in the dataset.

#### 04. Which player(s) had played for the longest period of time?

```python
# splitting the 'Span' column based on the "-"
data_span = data['Span'].str.split("-", expand=True)

# conccatinating the new dataframe with the main dataframe
data = pd.concat([data, data_span], axis=1)

# renaming the newly created column names
data = data.rename(columns={0: "Start Year",
                        1: "End Year"})

# removing the "Span" column
data = data.drop("Span", axis=1)

# Converting the object to int
data['Start Year'] = data['Start Year'].astype('int') 
data['End Year'] = data['End Year'].astype('int')

data['Years_Played'] = data['End Year'] - data['Start Year']

data = data.drop(['Start Year', "End Year"], axis=1)

# Highest number of years played
data.sort_values(by='Years_Played', ascending = False).head()
```

Imran Khan of Pakistan played highest (21) number of years.

#### 05. Which player(s) had played for the shortest period of time?

```python
# SHortest period of time
data.sort_values(by='Years_Played', ascending = True).head()
```

GP Swann of England played shortest (5 years) period of times.
